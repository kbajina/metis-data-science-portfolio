With an imbalanced dataset, the accuracy of predictions will be high because the desired result we are trying to predict (in this case applicant's who will default on their home credit) occurs in very few situations. For this type of problem we need to focus more on improving "recall" in order to capture as many instances of default as possible.

## Cost Breakdown:
1) **True Positive >>>** `Predict "At Risk" Accurately`
* Total number of home credit loan applicants that are accurately determined to be "At Risk" of Default
* The benefit of this would be filtering out applicants for pre-approval status:
    * Avoid loss on re-sale of the home through short-sale or auction:
    >– Short Sale (Foreclosure or Auction) results in 21% loss in home value <https://homeguides.sfgate.com/much-short-sale-lower-price-home-56901.html>  
    >— Avg. Credit Amount (based on dataset) = 598,552  
    > **`* THEREFORE, savings for accurately denying credit is: 125,695.92 (per applicant)`**
    * Avoid loss of future interest revenue:
    >— Assume most loands have 20% of principal paid off before default (5.5 years, based on avg. credit amount)
    >— Amounts to 120,701 paid off, with interest on the balance being the lost opportunity
    <https://www.bankrate.com/calculators/mortgages/mortgage-calculator.aspx>
    > **`* THEREFORE, opportunity cost of lost interest revenue is: 125,556.58 (per applicant)`**
    
2) **True Negative >>>** `Predict "No Risk" Accurately`
* Total number of home credit loan applicants that are accurately determined to have "No Risk" of Default
* The benefit of this would be the total number of customers that can be pre-approved:
    * Earned interest revenue:
    >— Avg. Credit Amount (based on dataset) = 598,552  
    >— Avg. Interest Rate = 3.75%  
    >— Assume 20 year mortgage  
    <https://www.bankrate.com/calculators/mortgages/mortgage-calculator.aspx>
    > **`* THEREFORE, the interest revenue gained would be: 253,200 (per applicant)`**  
    
3) **False Positive >>>** `Predict "At Risk" Inaccurately`
* Total number of home credit loan applicants that are innacurately determined to be "At Risk" of Default
* The cost of this would be customers who should have been pre-approved, but were not
    * Loss of interest revenue:
    >— Avg. Credit Amount (based on dataset) = 598,552  
    >— Avg. Interest Rate = 3.75%  
    >— Assume 20 year mortgage  
    <https://www.bankrate.com/calculators/mortgages/mortgage-calculator.aspx>
    > **`* THEREFORE, the lost potential interest revenue would be: 253,200 (per applicant)`** 
   
4) **False Negative >>>** `Predict "No Risk" Inaccurately`
* Total number of home credit loan applicants that are innacurately determined to have "No Risk" of Default
* The cost of this would be customers pre-approved for a loan but could potentially default:
    * Loss on re-sale of the home through short-sale or auction:
    >– Short Sale (Foreclosure or Auction) results in 21% loss in home value <https://homeguides.sfgate.com/much-short-sale-lower-price-home-56901.html>  
    >— Avg. Credit Amount (based on dataset) = 598,552  
    > **`* THEREFORE, cost of home re-sale is: 125,695.92 (per applicant)`**

## Optimize Threshold

```python
def net_benefit_loss(model, X, y, threshold):
    """
    Given a certain threshold value (float), determine the annual net benefit or loss
    derived from the classifier model for that threshold level for class seperation. 
    This will account for the benefits of True Postives & True Negatives, and the cost 
    of False Positive and False Negatives.
    
    Parameters
    ----------
        model: sklearn classifier
            The type of classifier model being used to predict y.
            
        X: array, shape = [n_samples]
            An array of independent variables.
            
        y: array, shape = [n_samples]
            An array of the dependent variable/outcome.
            
        threshold: float, (0,1) inclusive
            Indicates the class seperation based on the probability
            of the target outcome (1).
    """
    ## Make prediction based on defined threshold
    ## Create confusion matrix to determine truths and falses
    prediction = (model.predict_proba(X)[:,1] > threshold)
    matrix = confusion_matrix(y, prediction)
    
    ## Define values for True Postive, True Negative, False Positive, False Negative
    tp = matrix[1][1]
    tn = matrix[0][0]
    fp = matrix[0][1]
    fn = matrix[1][0]
    
    ## Define benefit or cost of each scenario
    tp_benefit = 125_695 + 125_556                # Avoid loan default (loss on sale + potential interest revenue)
    tn_benefit = 253_200                          # Revenue earned mortgage interest
    fp_cost = 253_200                             # Lost revenue on mortgage interest
    fn_cost = 125_695 + 125_556                   # Loan default (loss on sale + potential interest revenue)
    
    ## Calculate total benefit or loss
    result = (tp * tp_benefit) + (tn * tn_benefit) - (fp * fp_cost) - (fn * fn_cost)
    
    return round(result, 2)
```

```python
thresholds = defaultdict(int)

for val in [round(x, 2) for x in np.linspace(0.0, 1.0, 21)]:
    thresholds[val] = net_benefit_loss(logreg2, X_te, y_test, val)

#for key in thresholds:
#    thresholds[key] = np.round(thresholds[key]/thresholds[1], 4)

thresholds
```

## Observations

When trying to optimize the threshold value simply based the cost/benefit of True Positive, True Negative, False Postive and False Negative outcomes, it became clear that this is a numbers game. The fact that the proportion of Positive to Negative outcomes is so small, and that the losses incurred due to defaults are not substantially higher than the gains from interest revenue, the incentive is to simply provide a home credit loan to everyone. 

The metric used for the graph above looks at the benefit of each unit increase in the threshold value, relative to the net gain from setting a threshold of 1. The reason for this, is because the weights being applied to each scenario in the confusion matrix shouldn't be directly represented as actual profit or loss, as there are many other considerations required to assess that. This comparison simply looks at the basic opportunity costs of classifying any one applicant correctly or incorrectly as "At Risk of Default" or "Not at Risk of Default".

Also note, this analysis does not account for considerations such as banking regulations and the ethics behind predatory lending practices. For this reason, it makes sense to continue with the default threshold of 0.5. This threshold still provides a recall score of 73% while returning a net benefit of $730M over a 20 year period.

```python
# The scoring function that will use the Gradient Boosting Classifier to classify new data points
def DefaultRisk(applicant_years_employed,
            applicant_age_years,
            flag_own_car,
            current_cc_limit,
            applicant_completed_post_secondary,
            days_id_publish,
            amt_goods_price,
            cash_loan_max_future_installments,
            cash_loan,
            flag_work_phone,
            loan_beyond_purchase_price,
            region_rating_client,
            is_male,
            recent_cc_balance,
            annual_annuity_income_ratio):
    X = np.column_stack([applicant_years_employed,
            applicant_age_years,
            flag_own_car,
            current_cc_limit,
            applicant_completed_post_secondary,
            days_id_publish,
            amt_goods_price,
            cash_loan_max_future_installments,
            cash_loan,
            flag_work_phone,
            loan_beyond_purchase_price,
            region_rating_client,
            is_male,
            recent_cc_balance,
            annual_annuity_income_ratio])
    X = scaler.transform(X)
    return encoder.inverse_transform(logreg2.predict(X)).tolist()
```
